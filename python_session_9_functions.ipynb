```python
# First create 4 individual functions
# fun1: add
# fun2: sub
# fun3: mul
# fun4: div
# Create a main function name : calculator
# Inside main function
# print some functions 
# option 1: addition
# option 2: sub
# option 3: mul
# option 4: div

def add(a,b):
    print('answer:',a+b)
def sub(a,b):
    print('answer:',a-b)
def mul(a,b):
    print('answer:',a*b)
def div(a,b):
    print('answer:',a/b)
def calculator():
    print('option1=addition')
    print('option2=substraction')
    print('option3=multiplication')
    print('option4=Division')
    option=eval(input('enter the option 1/2/3/4:'))
    a=eval(input('enter a value'))
    b=eval(input('enter b value'))
    if option==1:
        add(a,b)
    elif option==2:
        sub(a,b)
    elif option==3:
        mul(a,b)
    elif option==4:
        div(a,b)
    else:
        print('enter a valid option')
        
calculator()
        


    
```
